```python
tickers = []
for i in df_SP500.Symbol:
  tickers.append(i)
```

Buscamos en la api los datos para nuestros tickers

```python
api_key = "XXXX"
```

```python
years = ['2021', '2020', '2019', '2018', '2017']
```

```python
list_fundamentals = []
for ticker in tickers:
  for year in years:
    list_row = []
    try:
      marketcap = fa.enterprise(ticker, api_key)[year].loc['marketCapitalization']
      net_income = fa.income_statement(ticker, api_key, period="annual")[year].loc['netIncome']
      revenue = fa.income_statement(ticker, api_key, period="annual")[year].loc['revenue']
      fcf = fa.cash_flow_statement(ticker, api_key, period="annual")[year].loc['freeCashFlow']
      ltl = fa.balance_sheet_statement(ticker, api_key, period="annual")[year].loc['totalNonCurrentLiabilities']
      so = fa.enterprise(ticker, api_key)[year].loc['numberOfShares']
      debt_equity = fa.balance_sheet_statement(ticker, api_key, period="annual")[year].loc['totalDebt'] + fa.balance_sheet_statement(ticker, api_key, period="annual")[year].loc['totalEquity']
      list_row.append(ticker)
      list_row.append(year)
      list_row.append(marketcap)
      list_row.append(net_income)
      list_row.append(revenue)
      list_row.append(fcf)
      list_row.append(ltl)
      list_row.append(so)
      list_row.append(debt_equity)
    except:
      True
    list_fundamentals.append(list_row)

print(list_fundamentals)
```

```python
import numpy as np
```

```python
lf = list_fundamentals
```

Eliminamos los valores donde no se pusieron numeros debido a que no existia la información (Pocos casos)

```python
for i in lf:
  if len(i) != 9:
    lf.remove(i)
```

```python
for i in lf:
  if len(i) != 9:
    print(i)
```

Transformamos todo a un data frame

```python
df2 = pd.DataFrame(np.array(lf),
                   columns=['ticker', 'year', 'marketcap', 'net_income', 'revenue', 'fcf', 'ltl', 'so', 'debt_equity'])
```
